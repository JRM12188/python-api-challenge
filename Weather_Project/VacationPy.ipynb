```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint

# Import API key
from api_keys import g_key
gmaps.configure(api_key=g_key)

outputFilePath = os.path.join("Results-Data")
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
Cities_Climate_df = pd.read_csv(os.path.join("Results-Data","Cities_Climate.csv"),delimiter=",")
Cities_Climate_df
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Pull the locations based on there Latitude and Longitude of each city
locations = Cities_Climate_df[["Latitude","Longitude"]].astype(float)

#Drop any NAN values in case they weren't removed already
Cities_Climate_df = Cities_Climate_df.dropna()

#Use humidity as the float
humidity = Cities_Climate_df["Humidity"].astype(float)
MaxHumidity = humidity.max()
```

```python
#create a map using humidity and a heatmap layer
heat = gmaps.figure()
heatlayer = gmaps.heatmap_layer(locations, weights = humidity, dissipating=False, max_intensity = MaxHumidity, point_radius = 3)

heat.add_layer(heatlayer)
heat
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
Vacation_df = Cities_Climate_df.loc[(Cities_Climate_df["Max Temp"]>70)&
                                    (Cities_Climate_df["Max Temp"]<80)&
                                   (Cities_Climate_df["Cloudiness"]==0)&
                                   (Cities_Climate_df["Wind Speed"]<10)].reset_index(drop=True)
Vacation_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = Vacation_df.loc[:,["Latitude","Longitude","City","Country"]]

HotelName = []
HAddress = []
HRating = []
HLat = []
Hlng = []

for index, row in hotel_df.iterrows():
    CurrentCity = row["City"]
    CurrentCountry = row["Country"]
    
    Target_Coord = f"{row['Latitude']},{row['Longitude']}"
    Target_Radius = 5000
    Type = "Hotel"
    keyword = "Hotel"
    
    #Set Paramaters
    params = {
        "location": Target_Coord,
        "radius": Target_Radius,
        "type": Type,
        "keyword": keyword,
        "key": g_key
    }
    
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response = requests.get(url,params=params).json()
    results = response['results']
    
    #print(results)
    
    try:
        HotelName.append(results[0]['name'])
        HAddress.append(results[0]['vicinity'])
        HRating.append(results[0]['rating'])
        HLat.append(results[0]['geometry']['location']['lat'])
        Hlng.append(results[0]['geometry']['location']['lng'])
        print(f"Process records for hotel in {CurrentCity}, {CurrentCountry}")
    except:
        HotelName.append(np.nan)
        HAddress.append(np.nan)
        HRating.append(np.nan)
        HLat.append(np.nan)
        Hlng.append(np.nan)
        print(f"No results for hotel in {CurrentCity}, {CurrentCountry}. Skipping...")
```

```python
# Add columns to hotel_df of the hotels information
hotel_df["Hotel Name"] = HotelName
hotel_df["Address"] = HAddress
hotel_df["Rating"] = HRating
hotel_df["Lat"] = HLat
hotel_df["Lng"] = Hlng

# remove any nan rows
hotel_df = hotel_df.dropna()
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
heat.add_layer(markers)
# Display figure
heat
```
